```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv

PATH = Service("/Users/gindychen/Desktop/chromedriver.exe")
driver = webdriver.Chrome(service = PATH)

driver.get("https://www.binghamton-ny.gov/home")  
gov = driver.find_element(By.XPATH, '//*[@id="dropdownrootitem3"]/a')
depart = driver.find_element(By.XPATH, '//*[@id="dropdownrootitem3"]/div/div/ul[1]/li/a')
actions = ActionChains(driver)
actions.move_to_element(gov)
actions.click(gov)
actions.click(depart)
actions.perform()
csv_file = open("job_info.csv", "w", newline="", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["JOB", "TYPE", "APPLICATION DEADLINE", "SALARY"]) # column names 

r= 1
try: 
    main = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="widget_4_33_127"]/ul/li[16]/a'))
    )
    main.click()
    main1 = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="leftNav_1038_0_145"]/ul/li/ul/li[14]/ul/li/a'))
    )
    main1.click()
        
    for i in driver.find_elements(By.XPATH,'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr'):
        job= i.find_element(By.XPATH,f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{r}]/td[1]/a').text
        types=i.find_element(By.XPATH,f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{r}]/td[2]').text
        deadline=i.find_element(By.XPATH, f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{r}]/td[3]').text
        salary= i.find_element(By.XPATH,f'//*[@id="ColumnUserControl4"]/div[2]/table/tbody/tr[{r}]/td[4]').text
        r += 1
        csv_writer.writerow([job, types, deadline, salary])
    csv_file.close()
finally :
    driver.quit()
```
