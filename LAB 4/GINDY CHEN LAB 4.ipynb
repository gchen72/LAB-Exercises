```python
from bs4 import BeautifulSoup
import requests
import html5lib 
import csv
import pandas as pd
```

# IMDB WEBSITE OF THE TOP 100 MOVIES


```python
source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup= BeautifulSoup(source,"lxml")
```

```python
section = soup.find("div", class_="lister-item-content") #isolating div & the class
print (section.prettify())
```

```python
csv_file = open("movie.csv", "w", newline="", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Movie Title", "Date", "Genre", "Rating"]) # column names 

for element in soup.find_all("div", class_="lister-item-content"):
    title = element.h3.a.text
    date= element.find(class_="lister-item-year text-muted unbold").text.strip("-")
    genre = element.find(class_="genre").text
    rating = element.find(class_="ipl-rating-star__rating").text
    csv_writer.writerow([title, date, genre, rating])  
    
csv_file.close()
```

```python
library = requests.get(f"https://www.loc.gov/search/?q=cats&sp=1.html").text
library_soup = BeautifulSoup(library, "lxml")
library_section = library_soup.find("div", class_="description") 
print (library_section.prettify())
# title, item description, webpage hyperlink
title = library_section.a.text.strip()
print (title)
item_description = library_section.find(class_="item-description-abstract").text.strip()
print (item_description)
```

```python
new_csv = open("library.csv", "w", newline ="", encoding = "utf-8")
csv_writer = csv.writer (new_csv)
csv_writer.writerow(["Title", "Item Description", "Webpage Hyperlink"])
page = 1
num = 1
for i in range(1,6):
    books = requests.get(f"https://www.loc.gov/search/?q=cats&sp={i}").text
    soup = BeautifulSoup(books, "lxml")
    for element in soup.find_all( class_="item-description"):
        title = element.a.text.strip()  # or use class="item-description-title"
        if element.find('span', class_="item-description-abstract") == None: 
            item_description == "None"
        else:
            item_description = element.find('span', class_="item-description-abstract").text.strip()
        link = element.a.get('href')
        csv_writer.writerow([title, item_description, link])
        item_description = "None"
new_csv.close()

```
