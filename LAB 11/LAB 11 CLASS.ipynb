```python
from tkinter import *
import random
import requests
import json
file = requests.get('https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy' )
json_file= file.json()
root = Tk()
root.title("Trivia")
root.geometry("800x1400")
quest_list = json_file 
points = 0
right_ques_num = 0
def selection():
    ans =(rad.get())
    print(ans) 
    if ans == correct_answer_index:
        label["text"]= "You're right!"
        global points
        points += 10
        points_["text"] = f"Points Earned : {points}"
        global right_ques_num 
        right_ques_num += 1
    else : 
        label["text"]= "You're wrong :( !"
    if len(quest_list) == 0:
        feedback = Label(root, text = f'You got {right_ques_num} out of 10 questions right! Congrats!')
        feedback.pack()
    
def show_question():
    number_of_questions = len(quest_list) - 1
    label["text"] = ""
    if len(quest_list) == 1:
        end_game = Label(root, text= f"This is the end of the game. In total you earn {points} points! Congrats!")
        right_ques_number["text"] = f"You got {right_ques_num} number of question."
    ques = random.randint(0,number_of_questions)
    quest =quest_list[ques]
    quest_list.remove(quest)
    category=quest['category']
    #explain = Label(root, text=f"This question is from the {category}", font = ("Arial", 10)).pack()
    question= quest["question"]
    global quest_show
    quest_show["text"]=  question
    global answers
    answers = []
    global correct_answer
    correct_answer = quest["correctAnswer"]
    answers.append(correct_answer)
    global radio
    radio = StringVar()
    incorrect_answers= quest["incorrectAnswers"]
    for i in incorrect_answers :
        answers.append(i)
    #https://note.nkmk.me/en/python-random-shuffle/
    random.shuffle(answers)
    global correct_answer_index
    for i in answers:
        if i == correct_answer: 
            correct_answer_index = answers.index(i)
    #print(correct_answer_index)
    if len(quest_list) == 9:
        global r1
        global r2
        global r3
        global r4
        global rad
        rad = IntVar()
        r1 = Radiobutton(root,text= answers[0], variable=rad, value = 0, command= selection)
        r1.pack()
        r2= Radiobutton(root,text= answers[1], variable= rad, value = 1, command= selection)
        r2.pack()
        r3 = Radiobutton(root, text= answers[2],variable= rad, value = 2, command= selection)
        r3.pack()
        r4 = Radiobutton(root, text= answers[3], variable= rad, value = 3, command= selection)
        r4.pack()
    else: 
        r1["text"]= answers[0]
        r2["text"]= answers[1]
        r3["text"]= answers[2]
        r4["text"]= answers[3]
            
        
title = Label(root, text="It's time for Trivia!!!", font = ("Arial", 20)).pack()
title2 = Label(root, text="ARE YOU READY", font = ("Arial", 20)).pack()
ruready= Button(root, text = "Click if you are ready?", command = show_question ).pack()
quest_show = Label(root)
quest_show.pack() 
label=Label(root)
label.pack()
right_ques_number = Label(root, text = "")
next_question = Button(root, text = "Click for the next question!", command = show_question).pack()
points_ = Label (root, text = "Points Earned : 0")
points_.pack()

        
root.mainloop()
```
