```python
# GUI # 1
import requests
from tkinter import *
r = requests.get("https://api.quotable.io/authors")
data = r.json()
data_list = data["results"]
a_list =[]
author=""
for i in range(0,len(data_list)):
    a=data_list[i]
    a_list.append(a["name"])

root = Tk()
root.title("QUOTE BY AUTHOR")
root.geometry("600x400")
def chosen_option():
    global author
    author = value.get()
    #result_label = Label(root, text=f"You've chosen {author}")
    #result_label.pack()
    getquote()
value = StringVar()
value.set("Select a AUTHOR")

dropdown = OptionMenu(root, value, *a_list)
choose_button = Button(root, text="Click to choose", command=chosen_option)

dropdown.pack()
choose_button.pack()

def getquote():
    text_box.delete("1.0","end")
    r = requests.get(f"https://api.quotable.io/quotes?author={author}")
    data = r.json()
    quote = data["results"][1]["content"]
    text_box.insert(END, quote)

intro = Label(root, text=f"Here is an inspirational quote for you.")
text_box = Text(root, height = 4, width = 30, bg="#D1E1F3", padx=30, pady=50)

intro.pack()
text_box.pack()

root.mainloop()
```
